```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
from matplotlib.pyplot import plot, savefig

# Variables
x_values = np.linspace(-20, 20, 5000)

# Length Containment function
def l(k):
    return  np.power(2.0, -k)*np.cos(k*np.pi)

# Polynomial for shifting
def h(x):
    return (x+6)*(x-2)**2 *0.25

# Initialize h_values array with the original function values
h_values = h(x_values)

k=6

# Shifting Function
for n in range(k):
    length = l(n)
    for i in range(-(n+1)*5000, (n+1)*5000):
        condition = np.logical_and((2 * i * length <= x_values), (x_values <= (2 * i + 1) * length))
        if condition.any():  # Check if any element in the condition array is True
            x_values[condition] -= 2 * length

# Sort x_values and rearrange h_values accordingly
sorting_indices = np.argsort(x_values)
x_values = x_values[sorting_indices]
h_values = h_values[sorting_indices]


# Define a function for the threshold based on the position in the array
def calculate_threshold(index):
    return 0.3  # Adjust the function as needed

# Detect and mark points with large discontinuities
differences = np.abs(np.diff(h_values))
thresholds = np.array([calculate_threshold(i) for i in range(1, len(h_values))])
discontinuity_indices = np.where(differences > thresholds)[0] + 1  # Adding 1 to get the correct indices

# Mark points with large discontinuities as np.nan
h_values[discontinuity_indices] = np.nan



# Plotting the final function
plt.plot(x_values, h_values, '-', color ='black', label='Shifted Set')
plt.xlim(-10, 7)
plt.ylim(0,20)
plt.title('Seed after '+ str(k)+ ' shifts, generalized as points')
plt.xlabel('x')
plt.ylabel('y')
#savefig('Shift.png', transparent=True)
#plt.legend()
plt.show()
```
