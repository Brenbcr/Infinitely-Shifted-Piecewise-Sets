```python
from sympy import *
from spb import *
```

```python
x = Symbol('x', real = True)
k = Symbol('k', natural = True)
```

```python
f = Function('f', real=True)(x)
f = (x)**2
```

```python
l = Function('l', natural=True)(k)
l = 2**(-k)
length = l.subs(k, 1)
```

```python
h = Piecewise ( *[ (f.subs(x, x + 2*length), (2*i*length <= x) & (x <= (2*i+1)*length)) for i in range(10) ], (f, True))
```

```python
plot_piecewise(h, xlim=(0,5), ylim=(0,15))
```

```python
plot(h, xlim=(0,5), ylim=(0,30))
```

```python
length1 = l.subs(k, 2)
h1 = Piecewise ( *[ (h.subs(x, x + 2*length1), (2*i*length1 <= x) & (x <= (2*i+1)*length1)) for i in range(10) ], (h, True))
```

```python
plot_piecewise(h1, xlim=(0,5), ylim=(0,25))
```

```python
plot_piecewise(h1, xlim=(0,5), ylim=(0,25))
```

```python
plot(h1, xlim=(0,5), ylim=(0,40))
```

```python
length2 = l.subs(k, 3)
h2 = Piecewise ( *[ (h1.subs(x, x + 2*length2), (2*i*length2 <= x) & (x <= (2*i+1)*length2)) for i in range(15) ], (h1, True))
```

```python
plot(h2, xlim=(0,5), ylim=(0,40))
```
